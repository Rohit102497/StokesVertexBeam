```python
import pickle
import matplotlib.pyplot as plt
```

```python
filename = "/StokesVertexBeam/results/Design1/densenet.data"
```

```python
# open a file, where you stored the pickled data
file = open(filename, 'rb')
# dump information to that file
data = pickle.load(file)
plt.plot(data['Train_loss'], label = 'Train Loss')
plt.plot(data['Val_loss'], label = 'Validation Loss')
#adding text inside the plot
plt.text(35, 0.75, 'Exp acc  = ' + str(round(data['experimental_acc'], 4)), fontsize = 10)
plt.text(35, .7, 'Test acc  = ' + str(round(data['test_acc'], 4)), fontsize = 10)
plt.text(35, 0.65, 'Val acc  = ' + str(round(data['Val_acc'][49].item(), 4)), fontsize = 10)
plt.text(35, .6, 'Train acc  = ' + str(round(data['Train_acc'][49].item(), 4)), fontsize = 10)
plt.legend()
plt.title("Alexnet, Exp 1, BS 8, lr 0.001, momentum 0.9, epochs 50" )
print(round(data['Train_loss'][49], 4), round(data['Val_loss'][49], 4))
print(round(data['experimental_acc'], 4), round(data['test_acc'], 4), 
    round(data['Val_acc'][49].item(), 4), round(data['Train_acc'][49].item(), 4))
```
