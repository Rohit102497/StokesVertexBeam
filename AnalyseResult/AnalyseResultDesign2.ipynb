```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = "/StokesVertexBeam/results/Design2/densenet.data"
```

```python
def comp_confmat(actual, predicted):

    # extract the different classes
    classes = np.unique(actual)

    # initialize the confusion matrix
    confmat = np.zeros((len(classes), len(classes)))

    # loop across the different combinations of actual / predicted classes
    for i in range(len(classes)):
        for j in range(len(classes)):

           # count the number of instances in each combination of actual / predicted classes
           confmat[i, j] = np.sum((actual == classes[i]) & (predicted == classes[j]))

    return confmat
```

```python
def analyse(data):
    actual_label = []
    prediction_label = []
    fold_list = ["1stFold", "2ndFold", "3rdFold", "4thFold", "5thFold", "6thFold", "7thFold", "8thFold", "9thFold", "10thFold"]

    acc = []
    for i in fold_list:
        # print(len(data[i]['exp_val_actual_label']))
        acc.append(np.round(data[i]['exp_val_acc'], 2))
        print(i, np.round(data[i]['exp_val_acc'], 2))
        actual_label = actual_label + data[i]['exp_val_actual_label']
        prediction_label = prediction_label + data[i]['exp_val_prediction']
    
    print("Toal acc ", np.mean(acc))
    print(len(actual_label), len(prediction_label))



    # sample data
    actual = [1, 3, 3, 2, 5, 5, 3, 2, 1, 4, 3, 2, 1, 1, 2]
    predicted = [1, 2, 3, 4, 2, 3, 3, 2, 1, 2, 3, 1, 5, 1, 1]

    # confusion matrix
    print(comp_confmat(actual_label, prediction_label))
```

```python
# open a file, where you stored the pickled data
file = open(filename, 'rb')
# dump information to that file
data = pickle.load(file)
analyse(data)
```

```python
# open a file, where you stored the pickled data
file_withoutSim = open(filename, 'rb')
# dump information to that file
data_withoutSim = pickle.load(file_withoutSim)
analyse(data_withoutSim)
```
